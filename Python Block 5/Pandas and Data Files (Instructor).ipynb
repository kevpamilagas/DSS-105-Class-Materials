```python
from datetime import datetime, timedelta

store1['DateTime'] = datetime(1900, 1, 1) + pd.to_timedelta(store1['Date'] + store1['Time'] - 2, unit='D')
store1['Month'] = store1['DateTime'].dt.month_name()
store1.head()
```

```python
o_num = 10000
a = store1['Time'][o_num]*2400

if a >= 1000 and a < 1100:
    print('Morning')
elif a >= 1100 and a < 1400:
    print('Lunch')
elif a >= 1400 and a < 1600:
    print('Mid-Day')
elif a >= 1600 and a < 2100:
    print('Dinner')
elif a >= 2100 and a <= 2200:
    print('Evening')
else:
    print('Error')

print(store1['DateTime'][o_num])
```

```python
def shift_name(time):
    if time < 1100/2400:
        return 'Morning'
    elif time < 1400/2400:
        return 'Lunch'
    elif time < 1600/2400:
        return 'Mid-Day'
    elif time < 2100/2400:
        return 'Dinner'
    else:
        return 'Evening'
```

```python
store1['Shift'] = store1['Time'].apply(shift_name)
store1.head()
```
